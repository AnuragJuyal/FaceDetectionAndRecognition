```python
import cv2
import numpy as np
import os
from sklearn.neighbors import KNeighborsClassifier as Knn
```

```python
model=cv2.CascadeClassifier("raw.githubusercontent.com_opencv_opencv_4.x_data_haarcascades_haarcascade_frontalface_alt.xml")
camera=cv2.VideoCapture(0)
offset=30
```

```python
dataset_path="./data/"
faceData=[]
labels=[]
nameMap={}
classId=0

for f in os.listdir(dataset_path):
    if f.endswith(".npy"):
        nameMap[classId]=f[:-4]
        dataItem=np.load(dataset_path+f)
        m=dataItem.shape[0]
        faceData.append(dataItem)
        target=classId*np.ones((m,))
        classId+=1
        labels.append(target)
X=np.concatenate(faceData,axis=0)
Y=np.concatenate(labels,axis=0).reshape((-1,1))
```

```python
mod=Knn(4)
mod.fit(X,Y)
while True:
    success,img=camera.read()
    if not success:
        print("Reading Camera Failed")
        break
    faces=model.detectMultiScale(img,1.3,2)
    print(faces)
    if len(faces)>0:
        for f in faces:
            x,y,w,h=f
            cropped_face=img[y-offset:y+h+offset,x-offset:x+w+offset]
            cropped_face=cv2.resize(cropped_face,(100,100))
            classPredicted=int(mod.predict(cropped_face.flatten().reshape(1,-1)))
            namePredicted=nameMap[classPredicted]
            cv2.putText(img,namePredicted,(x,y-10),cv2.FONT_HERSHEY_SIMPLEX,1,(250,250,0),2,cv2.LINE_AA)
            cv2.rectangle(img,(x,y),(x+w,y+h),(250,250,50),2)
    cv2.imshow("Prediction Window",img)
    key=cv2.waitKey(1)
    
    if key==ord('q'):
        break
camera.release()
cv2.destroyAllWindows()
```
